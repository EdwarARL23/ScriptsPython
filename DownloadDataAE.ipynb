```python
df_DemSIN= apiXM.request_data(pydataxm.ReadDB(),    #Se indica el objeto que contiene el serivicio
                        "DemaSIN",                #Se indica el nombre de la métrica tal como se llama en el campo metricID
                        "Sistema",                 #Campo que indica si es Sistema, Agente, Recurso, Comercializador
                        FechaIni,       #Corresponde a la fecha inicial de la consulta
                        FechaFin)      #Corresponde a la fecha final de la consulta
df_DemSIN.rename(columns={'Value':'DemSIN'},inplace=True)
df_DemSIN['DemSIN']=round(df_DemSIN['DemSIN']/1e6,2)
```

```python
df_GenRecdia=df_GenReal.copy()
df_GenRecdia['Total']=0

df_DemReal=df_DemRealIni.copy()
df_DemReal['TotalDem']=0

l_per=[]
for i in range(1,25):
    if i<10:
        l_per.append('Values_Hour0' + str(i))
    else:
        l_per.append('Values_Hour' + str(i))


for col in l_per:
    df_GenRecdia['Total']=df_GenRecdia['Total']+df_GenRecdia[col]
    df_DemReal['TotalDem']=df_DemReal['TotalDem']+df_DemReal[col]
# df_GenRecdia.to_csv('GenRe.csv')

df_GenRecdia=df_GenRecdia[['Values_code','Date','Total']]
df_DemReal=df_DemReal[['Date','TotalDem']]

df_GenRecdia['Total']=df_GenRecdia['Total']/1e6
df_DemReal['TotalDem']=df_DemReal['TotalDem']/1e6

df_GenRecdia=df_GenRecdia.fillna(0)
df_DemReal=df_DemReal.fillna(0)

df_GenRecdia=df_GenRecdia.merge(df_Rec,left_on=['Values_code'],right_on=['Values_Code'], how='left')[['Values_code','Date','Total','Values_Name', 'Values_Type', 'Values_Disp','Values_RecType', 'Values_State']]
df_GenRecdia.head(5)
# df_GenRecdia.to_csv('GenReal.csv')
```

```python
#Generación térmica
df_termia=df_GenRecdia[(df_GenRecdia.Values_Type=='TERMICA') & (df_GenRecdia.Values_Disp=='DESPACHADO CENTRALMENTE')]
df_termia=df_termia.groupby(['Date'])['Total'].sum()
df_termia=df_termia.reset_index()
df_termia.rename(columns={'Total':'Termica'},inplace=True)
# df_termia.head()
# df_termia.to_csv('Termica.csv')

#Generación térmica TPL
df_termiaTPL=df_GenRecdia[(df_GenRecdia.Values_Type=='TERMICA') & (df_GenRecdia.Values_Disp=='DESPACHADO CENTRALMENTE') & 
                          (df_GenRecdia['Values_Name'].isin(['TEBSAB CC','BARRANQUILLA 3','BARRANQUILLA 4','TERMOCANDELARIA CC']))]
df_termiaTPL=df_termiaTPL.groupby(['Date'])['Total'].sum()
df_termiaTPL=df_termiaTPL.reset_index()
df_termiaTPL.rename(columns={'Total':'TermicaTPL'},inplace=True)


#Generación Hidráulica
df_hidro=df_GenRecdia[(df_GenRecdia.Values_Type=='HIDRAULICA') & (df_GenRecdia.Values_Disp=='DESPACHADO CENTRALMENTE') & (~df_GenRecdia['Values_RecType'].isin(['FILO AGUA ESPECIAL','FILO DE AGUA']))]
# df_hidro.to_csv('Hidro.csv')
df_hidro=df_hidro.groupby(['Date'])['Total'].sum()
df_hidro=df_hidro.reset_index()
df_hidro.rename(columns={'Total':'Hidro'},inplace=True)

#Generación Hidráulica filo agua DC
df_hidroFA=df_GenRecdia[(df_GenRecdia.Values_Type=='HIDRAULICA') & (df_GenRecdia.Values_Disp=='DESPACHADO CENTRALMENTE') & (df_GenRecdia['Values_RecType'].isin(['FILO AGUA ESPECIAL','FILO DE AGUA']))]
# df_hidro.to_csv('Hidro.csv')
df_hidroFA=df_hidroFA.groupby(['Date'])['Total'].sum()
df_hidroFA=df_hidroFA.reset_index()
df_hidroFA.rename(columns={'Total':'HidroFA'},inplace=True)

#Generación NDC
df_ndc=df_GenRecdia[(df_GenRecdia.Values_Disp=='NO DESPACHADO CENTRALMENTE')]
df_ndc=df_ndc.groupby(['Date'])['Total'].sum()
df_ndc=df_ndc.reset_index()
df_ndc.rename(columns={'Total':'NDC'},inplace=True)

#Generación otros DC
df_odc=df_GenRecdia[(~df_GenRecdia['Values_Type'].isin(['TERMICA','HIDRAULICA'])) & (df_GenRecdia.Values_Disp=='DESPACHADO CENTRALMENTE')]
df_odc=df_odc.groupby(['Date'])['Total'].sum()
df_odc=df_odc.reset_index()
df_odc.rename(columns={'Total':'OtroDC'},inplace=True)
```

```python
df_data=df_VolumenU.copy()
df_data=df_data.merge(df_Aportes,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'VolUtil', 'Aportes']]
df_data=df_data.merge(df_AporMH,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolUtil', 'Aportes']]
df_data=df_data.merge(df_VolUPor,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes']]
df_data=df_data.merge(df_Vert,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes','Vert']]
df_data=df_data.merge(df_termia,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes','Vert','Termica']]
df_data=df_data.merge(df_termiaTPL,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes','Vert','Termica','TermicaTPL']]
df_data=df_data.merge(df_hidro,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes','Vert','Termica','TermicaTPL','Hidro']]
df_data=df_data.merge(df_ndc,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes','Vert','Termica','TermicaTPL','Hidro','NDC']]
df_data=df_data.merge(df_hidroFA,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes','Vert','Termica','TermicaTPL','Hidro','HidroFA','NDC']]
df_data=df_data.merge(df_odc,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes','Vert','Termica','TermicaTPL','Hidro','HidroFA','NDC','OtroDC']]
# df_data=df_data.merge(df_DemReal,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes','Vert','Termica','TermicaTPL','Hidro','HidroFA','NDC','OtroDC','TotalDem']]
df_data=df_data.merge(df_DemSIN,left_on=['Date'],right_on=['Date'], how='left')[['Date', 'ApoMH','VolPor','VolUtil', 'Aportes','Vert','Termica','TermicaTPL','Hidro','HidroFA','NDC','OtroDC','DemSIN']]
df_data=df_data.fillna(0)
df_data.head()
```
