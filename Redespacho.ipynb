```python
import datetime
import pandas as pd
import datetime as dt

encabezado=['Recurso']
Periodos=[]

encabezado2=['Variable','Descp']
encabezado3=['Variable']

for per in range(1,25):
    encabezado.append('P' + "{:02d}".format(per))
    encabezado2.append('P' + "{:02d}".format(per))
    encabezado3.append('P' + "{:02d}".format(per))
    Periodos.append('P' + "{:02d}".format(per))
#encabezado.append('Fecha')

fechaIni='2024-07'
fechFin='2024-08'
#fechFin='2018-11'

mes_ini = int(fechaIni[5:7:1])
mes_fin = int(fechFin[5:7:1])
ano_ini = int(fechaIni[0:4:1])
ano_fin = int(fechFin[0:4:1])


fecha_ini = datetime.datetime(ano_ini, mes_ini, 1)
fecha_fin = datetime.datetime(ano_fin, mes_fin, 19)
#fecha_fin = datetime.datetime(ano_fin, mes_fin, 2)
day = datetime.timedelta(days=1)

#df_Despacho=pd.DataFrame(columns=encabezado)

Lt_Despacho=[]
Lt_Redespacho=[]
Lt_rMar=[]
Lt_PBNA=[]

ConsultarPB=0

while fecha_ini <= fecha_fin:
    
    print(fecha_ini)
    ano = fecha_ini.year
    mes=fecha_ini.month
    dia=fecha_ini.day
    fecha_ini=fecha_ini+day
    
    ruta=r"C:\Información XM\Publico\Despacho\\"
    ruta1=r"C:\Información XM\Publico\Redespacho\\"

    ruta2=r"C:\Información XM\PUBLICOK\SIC\COMERCIA\\"
    
    # ruta="\\\cndras01\Publico\DESPACHO\\"
    # ruta1="\\\cndras01\Publico\Redespacho\\"
    
    ruta= ruta + str(ano) + '-' + "{:02d}".format(mes) + '\\'
    ruta= ruta + 'dDEC' + "{:02d}".format(mes) + "{:02d}".format(dia) + '.txt'
    
    rutafechar=ruta1 + str(ano) + '-' + "{:02d}".format(mes) + '\\'
    ruta1=rutafechar + 'rDEC' + "{:02d}".format(mes) + "{:02d}".format(dia) + '.txt'
    ruta12=rutafechar + 'rMAR' + "{:02d}".format(mes) + "{:02d}".format(dia) + '.txt'

    ruta2=ruta2 + str(ano) + '-' + "{:02d}".format(mes) + '\\'
    ruta2=ruta2 + 'trsd' + "{:02d}".format(mes) + "{:02d}".format(dia) + '.tx1'
    
    #print(ruta)    
    
    df_aux = pd.read_csv(ruta, sep=',', header=None,encoding="ISO-8859-1")
    df_aux.columns=encabezado
    df_aux=df_aux[df_aux.Recurso!='Total']
    df_aux['Fecha']=str(ano) + '-' + "{:02d}".format(mes) + '-' + "{:02d}".format(dia)
    df_aux['key']=df_aux['Recurso'] + str(ano) + "{:02d}".format(mes) + "{:02d}".format(dia) 
    df_aux.set_index('key',inplace=True) 
    df_aux.sort_index(inplace=True)
    Lt_Despacho.append(df_aux)
    
    df_aux = pd.read_csv(ruta1, sep=',', header=None,encoding="ISO-8859-1")
    df_aux.columns=encabezado
    df_aux=df_aux[df_aux.Recurso!='Total']
    df_aux['Fecha']=str(ano) + '-' + "{:02d}".format(mes) + '-' + "{:02d}".format(dia) 
    df_aux['key']=df_aux['Recurso'] + str(ano) + "{:02d}".format(mes) + "{:02d}".format(dia) 
    df_aux.set_index('key',inplace=True)
    df_aux.sort_index(inplace=True)
    Lt_Redespacho.append(df_aux)

    df_aux = pd.read_csv(ruta12, sep=',', header=None,encoding="ISO-8859-1")
    df_aux.columns=encabezado3
    df_aux['Fecha']=str(ano) + '-' + "{:02d}".format(mes) + '-' + "{:02d}".format(dia) 
    Lt_rMar.append(df_aux)

if ConsultarPB==1:
    if fecha_ini<pd.to_datetime('2024-02-19'):
        df_aux = pd.read_csv(ruta2, sep=';', header=None,encoding="ISO-8859-1")
        df_aux.columns=encabezado2
        df_aux=df_aux[df_aux.Variable=='PBNA']
        df_aux['Fecha']=str(ano) + '-' + "{:02d}".format(mes) + '-' + "{:02d}".format(dia) 
        Lt_PBNA.append(df_aux)
```

```python
df_Despacho=pd.concat(Lt_Despacho)
df_Redespacho=pd.concat(Lt_Redespacho)
df_rMar=pd.concat(Lt_rMar)

if ConsultarPB==1:
    df_PB=pd.concat(Lt_PBNA)
```

```python
RecursoFecha=df_Despacho[['Recurso','Fecha']]

df_Diferencia=df_Despacho[Periodos].subtract(df_Redespacho[Periodos])
```

```python
df_Diferencia['Total']=0

for col in Periodos:
    df_Diferencia['Total']=df_Diferencia['Total']+df_Diferencia[col]

df_Diferencia[['Recurso','Fecha']]=RecursoFecha

index_names0 = df_Diferencia[ df_Diferencia['Total'] == 0 ].index

df_Diferencia.drop(index_names0, inplace = True)

index_names0 = df_Diferencia[ df_Diferencia['Total'] > 0 ].index

df_Dif_g_0=df_Diferencia.loc[index_names0]

df_GroupDayg0=df_Dif_g_0.groupby(['Fecha']).sum()

index_names0 = df_Diferencia[ df_Diferencia['Total'] < 0 ].index

df_Dif_l_0=df_Diferencia.loc[index_names0]
```

```python
df_Variable=df_Dif_g_0[((df_Dif_g_0.Recurso=='AMOYA') | (df_Dif_g_0.Recurso=='RN_GUAYEPO') | (df_Dif_g_0.Recurso=='CUCUANA') 
                                       | (df_Dif_g_0.Recurso=='RN_PETALMGD') | (df_Dif_g_0.Recurso=='C_LLERAS') | (df_Dif_g_0.Recurso=='RN_CAICIENA') 
                                       | (df_Dif_g_0.Recurso=='DARIOVS')  
                                       | (df_Dif_g_0.Recurso=='ESCUELAMINAS') | (df_Dif_g_0.Recurso=='RN_SELPASO')
                                       | (df_Dif_g_0.Recurso=='ESMERALDA') | (df_Dif_g_0.Recurso=='SALTO II') 
                                       | (df_Dif_g_0.Recurso=='SANFRANCISCO') 
                                       | (df_Dif_g_0.Recurso=='SANMIGUEL'))]
```

```python
import holidays

l_holy=[]

for ano in [2023,2024]:
    # Get Colombia holidays for the year 2023 and 2024
    colombia_holidays = holidays.Colombia(years=ano)

    # Print the holidays
    for date, name in sorted(colombia_holidays.items()):
        #print(f"{date}: {name}")
        l_holy.append(pd.to_datetime(date))
```

```python
grupo=1
semanas=1
df_Variable['Clas']=''


rsize=df_Variable.shape[0]
indice=0
for ind in df_Variable.index:
    numsem=0
    fecha=df_Variable.at[ind,'Fecha']
    fecha=dt.datetime.strptime(fecha,'%Y-%m-%d')

    if indice<rsize-1:
        fech_ahead=df_Variable.iloc[indice+1,26]
        fech_ahead=dt.datetime.strptime(fech_ahead,'%Y-%m-%d')
        dayw=fech_ahead.weekday()


    if fecha in l_holy:
        df_Variable.loc[ind,'Clas']=str(grupo) + 'F'
        if fecha.weekday()==6 and dayw!=6:
            numsem=numsem+1

    elif fecha.weekday()<5:
        df_Variable.loc[ind,'Clas']=str(grupo) + 'O'
    elif fecha.weekday()==5:
        df_Variable.loc[ind,'Clas']=str(grupo) + 'S'
    else:
        df_Variable.loc[ind,'Clas']=str(grupo) + 'F'

        if dayw!=6:
            numsem=numsem+1
    
    if numsem>=semanas:
        grupo=grupo+1
    
    indice=indice+1
```

```python
df_Data=df_Variable.groupby(['Fecha','Clas'],as_index=False)[Periodos].sum()
```

```python
df_DataPro=df_Data.groupby(['Clas'],as_index=False)[Periodos].mean()
```
